```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as linear_model
```

```python
df=pd.read_csv("homeprice.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area(s)')
plt.ylabel('prices(us$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area',fontsize=20)
plt.ylabel('price',fontsize=20)
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color="blue")
```

```python
ca=pd.read_csv("canada_per_capita_income.csv")
ca
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(ca.year,ca["per capita income (US$)"],color='green',marker='*')
```

```python
reg2=linear_model.LinearRegression()
reg2.fit(ca[['year']],ca[['per capita income (US$)']])
```

```python
reg2.predict([[2023]])
```

```python
%matplotlib inline
plt.xlabel('year', fontsize=20)
plt.ylabel('per capita income (US$)', fontsize=20)
plt.scatter(ca.year, ca['per capita income (US$)'], color='green', marker='*')
plt.plot(ca.year, reg2.predict(ca[['year']]), color="red")
```
